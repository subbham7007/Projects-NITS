```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv("/home/subham/Documents/data_analytics/glass+identification/glass.data")
data = data.drop(['Id'],axis=1)
data.tail()
```

1. Boxplot

```python
rows = 5
cols = 2
count = 1
plt.figure(figsize=(18,8))
sns.set(style = 'white') 
for i in data.columns:
    ax = plt.subplot(cols,rows,count)
    plt.boxplot(data[i])
    plt.title(i)
    count += 1
plt.show()
```

2. Violinplot

```python
rows = 5
cols = 2
count = 1
plt.figure(figsize=(18,8))
sns.set(style = 'whitegrid') 
for i in data.columns:
    ax = plt.subplot(cols,rows,count)
    sns.violinplot(data = data[i])
    plt.title(i)
    count += 1
plt.show()
```

3. Violinplot according to every type of glass

```python
plt.figure(figsize=(18,20))
sns.set(style = 'whitegrid') 
count = 1
for i in data.columns[:len(data.columns)-1]:
    ax = plt.subplot(5,2,count)
    sns.violinplot(x="Type of glass", y=i, data=data)
    count += 1
plt.show()
```

4. Scatterplot

```python
plt.figure(figsize=(20,8))
sns.set(style = 'white') 
count = 1
for i in data.columns[:len(data.columns)-1]:
    ax = plt.subplot(2,5,count)
    sns.scatterplot(x="Type of glass", y=i, data=data)
    count += 1
plt.show()
# sns.scatterplot(x="Type of glass", y='Mg', data=data)
```

5. histogram

```python
plt.figure(figsize=(20,12))
count = 1
for i in data.columns:
    ax = plt.subplot(3,4,count)
    plt.hist(data[i], bins=25)
    plt.title(i)
    count += 1
plt.show()
```

6. correlation coeffecient

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

7. Pairplot

```python
cd = data[['Na','Mg','Al','Ba','Type of glass']]
plt.figure(figsize=(24,24))
sns.set(style = 'white') 
sns.pairplot(cd)
plt.show()
```

8. pairplot according to different classes

```python
sns_plot = sns.pairplot(cd, kind="scatter", hue="Type of glass", palette="Set2")
```

9. Barchart of Type of glass

```python
fr = data['Type of glass'].value_counts()
plt.bar(fr.index, fr)
plt.xlabel('Type of glass')
plt.ylabel("frequency")
plt.show()
```

10. Cumulative Distribution Function

```python
plt.figure(figsize=(18,12))
sns.set(style = 'whitegrid') 
ct = 1
for i in data.columns:
    ax = plt.subplot(3,4,ct)
    count, bins_count = np.histogram(data[i].values, bins=20)
    pdf = count / sum(count) 
    cdf = np.cumsum(pdf)
    plt.plot(bins_count[1:], cdf)
    plt.title(i)
    ct += 1
plt.show()
```
